**Group By And Aggregation**

```python
df =pd.read_csv(r"D:\Downloads\New Downloads\Pandas.Project\Flavors.csv")
df
```

```python
df.groupby('Base Flavor')
```

```python
group_by_frame=df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','sum','count']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','sum','count'], 'Texture Rating':['mean','sum','count']})
```

```python
df.groupby
```

```python
df
```

```python
df.groupby(['Base Flavor','Liked']).mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','sum','count']})
```
